# Método de Briggs para el cálculo de logaritmos decimales

Implementación del método histórico de Henry Briggs usando
5 raíces cuadradas redondeadas a 6 cifras decimales.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha=1/np.log(10)
```

```python
def log10_briggs(x,n=5):
    y=x
    for _ in range(n):
        y=np.round(np.sqrt(y),6)
    r=y-1
    return (2**n)*alpha*r
```

```python
xs=np.arange(1,11)
aprox=[]
real=[]
error_rel=[]

for x in xs:
    if x==1:
        aprox.append(0)
        real.append(0)
        error_rel.append(0)
    else:
        a=log10_briggs(x)
        r=np.log10(x)
        aprox.append(a)
        real.append(r)
        error_rel.append(abs(a-r)/abs(r))
```

```python
for i,x in enumerate(xs):
    print(x,aprox[i],error_rel[i])
```

```python
r=np.linspace(0,0.05,100)
plt.plot(r,np.log10(1+r),label="log10(1+r)")
plt.plot(r,alpha*r,label="αr")
plt.legend()
plt.grid()
plt.show()
```

```python
error_abs=[abs(aprox[i]-real[i]) for i in range(len(xs))]
print("Error absoluto máximo:",max(error_abs))
```
