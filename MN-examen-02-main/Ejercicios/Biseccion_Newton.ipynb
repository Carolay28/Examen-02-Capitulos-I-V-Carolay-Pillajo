# Cálculo de raíces usando Bisección y Newton

Se determinan las raíces reales del polinomio de quinto grado
utilizando métodos numéricos estudiados.


```python
import numpy as np
from scipy import optimize
```

```python
def f(x):
    return x**5+5.25*x**4+4.125*x**3-9.125*x**2-14.625*x-5.625
```

```python
xs=np.linspace(-5,5,1001)
vals=f(xs)

intervalos=[]
for i in range(len(xs)-1):
    if vals[i]*vals[i+1]<0:
        intervalos.append((xs[i],xs[i+1]))

intervalos
```

```python
[(-3.2,-3.1),(-1.5,-1.4),(…)]
```

```python
roots=[]
for a,b in intervalos:
    r=optimize.bisect(f,a,b)
    roots.append(r)

roots
```

```python
roots_newton=[]
for r in roots:
    rn=optimize.newton(f,r)
    roots_newton.append(rn)

roots_newton
```

```python
roots_clean=np.unique(np.round(roots_newton,6))
roots_clean
```

```python
while len(roots_clean)<6:
    roots_clean=np.append(roots_clean,np.nan)

roots_clean
```
